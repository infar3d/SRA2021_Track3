```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.io import fits

observation_path = '../../lab1_resources/ogg0m406-kb27-20210612-0230-s91.fits.fz'

observation = fits.open(observation_path)
image = observation['SCI'].data
uncertainty = observation['ERR'].data
```

```python
from matplotlib.colors import LogNorm

fig, ax = plt.subplots()
plt.title('"data" of an LCO image')
pos = ax.imshow((image)[950:1050, 1500:1600], norm=LogNorm(), aspect='auto')
fig.colorbar(pos, ax=ax)
plt.show()

fig, ax = plt.subplots()
plt.title('"uncertainty" of an LCO image')
pos = ax.imshow((uncertainty)[950:1050, 1500:1600], norm=LogNorm(), aspect='auto')
fig.colorbar(pos, ax=ax)
plt.show()

fig, ax = plt.subplots()
plt.title('Signal to noise of an LCO image')
pos = ax.imshow((image/uncertainty)[950:1050, 1500:1600], norm=LogNorm(), aspect='auto')
fig.colorbar(pos, ax=ax)
plt.show()
```
