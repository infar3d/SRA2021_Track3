```python
import rebound
import numpy as np
import matplotlib.pyplot as plt
from astropy import units as u

def to_radians(value):
    return (value*u.deg).to(u.rad).value

sim = rebound.Simulation()
sim.units = ('yr', 'AU', 'Msun') # set our units to years, astronomical units, and solar masses.
# lets initialize an alien solar system. The central star will be 1 solar mass, and 
# the companion will be 1/10 of that (about 100 jupiter masses). This is a low mass stellar companion, not a planet.
sim.add(m=1.)
sim.add(m=0, a=1., omega=to_radians(0), Omega=to_radians(0), e=0, inc=to_radians(0))

plt.figure()
fig = rebound.OrbitPlot(sim, unitlabel="[AU]", periastron=True, lw=2)
plt.title('The simplest plot you can make of an orbit. This is what is called a face-on orbit.')
plt.show()
```

```python

sim = rebound.Simulation()
sim.units = ('yr', 'AU', 'Msun') # set our units to years, astronomical units, and solar masses.
# lets initialize an alien solar system. The central star will be 1 solar mass, and 
# the companion will be 1/10 of that (about 100 jupiter masses). This is a low mass stellar companion, not a planet.
planet_masses = 0.0001

sim.add(m=1.)
sim.add(m=planet_masses, a=1., omega=to_radians(0), Omega=to_radians(0), e=0, inc=to_radians(0))
sim.add(m=planet_masses, a=1., omega=to_radians(30), Omega=to_radians(0), e=0, inc=to_radians(60))

# integrating forward in time:

sim.move_to_com()
step_in_time = 0.01
current_time = 0
for i in range(2):
    current_time += step_in_time
    sim.integrate(current_time)
    

plt.figure()
fig = rebound.OrbitPlot(sim, unitlabel="[AU]", periastron=True, lw=2)
plt.title('Some time later')
plt.show()
```
